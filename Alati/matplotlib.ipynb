```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Da bi koristili matplotlib moramo imati neke podatke koje ćemo vizuelizirati. Pripremili smo dataset sa <a href="finance.yahoo.com">Yahoo Finance</a>

```python
data = pd.read_csv('dataSet/MSFT.csv')
```

```python
data.head()
```

Najlakši način kako prikazati finansijske podatke jeste putem linijskog dijagrama što ćemo i učiniti. Ali prije nego uradimo to putem matplot liba to ćemo uraditi sa pandasom jer i on ima neke vizualizacijske funkcije

pandas plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

U kombinaciji sa pandas plot funkcijom možemo koristiti i matplotlib jer je pandas plot skraćena verzija matplotlib-a, te jako dobro rade zajedno, s tim u vezi možemo da uljepšavamo graf po našim ukusima.

```python
data.plot(kind="line", x="Date", y="Open", figsize=(10,8)) # kind je vrsta dijagrama, figsize
plt.xticks(rotation=45) # rotiramo datume
plt.ylabel("Cijena dionice", fontsize = 12)
plt.xlabel("Datum", fontsize = 12)
plt.title("Cijena Microsoftove dionice", fontsize = 12, fontweight = "bold")
```
